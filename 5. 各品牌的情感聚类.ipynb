```python
import pandas as pd # 以pd别名导入Python数据分析模块
import numpy as np # 导入数值计算模块
```

# 苹果情感聚类
## apple_pos

```python
apple = pd.read_csv('./review/apple.csv')
```

```python
apple_pos = apple[apple['sentiment'].str.contains('pos')]

# 重新编号
apple_pos = apple_pos.reset_index(drop=True)
apple_pos
```

```python
apple_pos.to_csv('./sentiment/apple_pos.csv', index = False, encoding = 'utf-8')
```

## apple_neg

```python
apple_neg = apple[apple['sentiment'].str.contains('neg')]

# 重新编号
apple_neg = apple_neg.reset_index(drop=True)
apple_neg
```

```python
apple_neg.to_csv('./sentiment/apple_neg.csv', index = False, encoding = 'utf-8')
```
