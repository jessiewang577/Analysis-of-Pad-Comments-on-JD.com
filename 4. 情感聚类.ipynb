```python
import pandas as pd # 以pd别名导入Python数据分析模块
import numpy as np # 导入数值计算模块
```

```python
reviews = pd.read_csv('./review/reviews_extracted.csv')
```

# 正面评论

```python
pos = reviews[reviews['sentiment'].str.contains('pos')]

#重新编号
pos = pos.reset_index(drop=True)
pos
```

```python
pos.to_csv('./sentiment/pos.csv', index = False, encoding = 'utf-8')
```

```python
# 仅选取正面分词
pos_cutword = pos["cutword"]
pos_cutword.to_csv('./sentiment/pos_cutword.csv', index = False, header = None, encoding = 'utf-8')
```

# 负面评论

```python
neg = reviews[reviews['sentiment'].str.contains('neg')]

#重新编号
neg = neg.reset_index(drop=True)
neg
```

```python
neg.to_csv('./sentiment/neg.csv', index = False, encoding = 'utf-8')
```

```python
# 仅选取负面分词
neg_cutword = neg["cutword"]
neg_cutword.to_csv('./sentiment/neg_cutword.csv', index = False, header = None, encoding = 'utf-8')
```
