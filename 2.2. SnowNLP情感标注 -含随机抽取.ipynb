```python
# 测试模型效果，导入测试数据
import pandas as pd
test1000 = pd.read_excel('./train/content_train_mark - 1000.xlsx')
test1000
```

```python
from snownlp import SnowNLP
from snownlp import sentiment
```

```python
s = []
for c in test1000['评论']:
    score = SnowNLP(c).sentiments
    if score>=0.5:
        s.append("pos")
    else:
        s.append("neg")
count = np.sum(s == test1000['情感'])
print('准确率为：',count/len(s))
```

```python
confusion_matrix = pd.crosstab(test1000['情感'], s, margins=True)  # 制作混淆矩阵
confusion_matrix
```

```python
print('基于SnowNLP的情感分析的准确率为：',(confusion_matrix.iat[0,0] + confusion_matrix.iat[1,1])/confusion_matrix.iat[2,2])
```
