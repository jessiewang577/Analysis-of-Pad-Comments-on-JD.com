```python
import pandas as pd # 以pd别名导入Python数据分析模块
import re # 导入正则表达式模块，可进行模式匹配、搜索和替换
import numpy as np # 导入数值计算模块
from sklearn.model_selection import train_test_split
```

```python
reviews = pd.read_csv('./review/reviews_cut.csv')
```

# 苹果评论

```python
apple = reviews[reviews['商品名称'].str.contains('iPad')]

#重新编号
apple = apple.reset_index(drop=True)
apple
```

```python
apple.to_csv('./review/apple.csv', index = False, encoding = 'utf-8')
```

```python
# 仅选取苹果的分词
applectwd = apple["cutword"]
applectwd.to_csv('./review/apple_cutword.csv', index = False, header = None, encoding = 'utf-8')
```

```python
# 按市场占比随机抽取苹果的评论

# 将特征（数据）划分到 X 中，标签划分到 Y 中
x = apple[["商品名称", "评价星级", "评论", "cutword"]]
y = apple["sentiment"]
# 使用train_test_split函数划分数据集(训练集占xx%，测试集(此处为市场占比)占82.35%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.8235, random_state=0)
```

```python
apple_test = x_test
apple_test["sentiment"] = y_test
apple_test
```

```python
# apple_test.to_csv('./review/apple_test.csv', index = False, encoding = 'utf-8')
```

# 小米评论

```python
xiaomi = reviews[reviews['商品名称'].str.contains('小米')]

#重新编号
xiaomi = xiaomi.reset_index(drop=True)
xiaomi
```

```python
xiaomi.to_csv('./review/xiaomi.csv', index = False, encoding = 'utf-8')
```

```python
# 仅选取小米分词
xiaomictwd = xiaomi["cutword"]
xiaomictwd.to_csv('./review/xiaomi_cutword.csv', index = False, header = None, encoding = 'utf-8')
```

```python
# 按市场占比随机抽取xiaomi的评论

# 将特征（数据）划分到 u 中，标签划分到 v 中
u = xiaomi[["商品名称", "评价星级", "评论", "cutword"]]
v = xiaomi["sentiment"]
# 使用train_test_split函数划分数据集(训练集占xx%，测试集(此处为市场占比)占41.80%)
u_train, u_test, v_train, v_test = train_test_split(u, v, test_size=0.3932, random_state=0)
```

```python
xiaomi_test = u_test
xiaomi_test["sentiment"] = v_test
xiaomi_test
```
